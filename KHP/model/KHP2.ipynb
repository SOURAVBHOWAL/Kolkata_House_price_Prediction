```python
import pandas as pd
import numpy as np
df1=pd.read_csv('KOLKATA_HOUSE_PRICE_2.csv')
df1
```

```python
df1["BHK"].unique()
```

```python
df2=df1.drop(['Flat_Price','EMI_Starts'],axis='columns')
```

```python
df2
```

```python
df2['Price_per_sq.ft']=df2['Price_per_sq.ft'].apply(lambda x: x.split('/')[0])
```

```python
dummies=pd.get_dummies(df2.Location)
```

```python
df3=pd.concat([df2,dummies],axis='columns')
df3['EMI']=df3['EMI'].apply(lambda x: x if x!='#VALUE!' else None)
df3=df3.dropna()
df3['Price_per_sq.ft']=df3['Price_per_sq.ft'].astype(float)
df3['EMI']=df3['EMI'].astype(float)
df3['BHK']=df3['BHK'].astype(float)
df3['Total_sq.ft']=df3['Total_sq.ft'].astype(float)
df3['Flat_Price_K']=df3['Flat_Price_K'].astype(float)
```

```python
df3['Total_sq.ft'].unique()
```

```python
y=df3.Flat_Price_K
x=df3.drop(['Flat_Price_K','Location','Price_per_sq.ft'],axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression
model1= LinearRegression()
model1.fit(X_train,y_train)
model1.score(X_test,y_test)
```

```python
model1.predict(x[1:4])
```

```python
def predict_price(location,emi,bhk,total_sq_ft): 
    a = np.zeros(len(x.columns))
    a[0] = bhk
    a[1] = total_sq_ft
    a[2] = emi
    
    for i in range(3,len(x.columns)):
        if(location==x.columns[i]):
            a[i] = 1
            return model1.predict([a])*1000
            break

    
    

    
```

```python
predict_price('Madhyamgram',27.54,4,900)
```
